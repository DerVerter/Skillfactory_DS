```python
import pandas as pd
```

```python
dates_df = pd.read_csv('data/dates.csv')
movies_df = pd.read_csv('data/movies.csv')
ratings1_df = pd.read_csv('data/ratings1.csv')
ratings2_df = pd.read_csv('data/ratings2.csv')
```

```python
import pandas as pd
```

```python
dates_df = pd.to_datetime(dates_df)
```

pd.concat()

```python
ratings_df = pd.concat(
    [ratings1_df, ratings2_df],
    ignore_index=True)
display(ratings_df)
```

```python
print('Число строк в таблице ratings: ', ratings_df.shape[0])
print('Число строк в таблице dates: ', dates_df.shape[0])
print(ratings_df.shape[0] == dates_df.shape[0])
```

```python
display(ratings1_df.tail(1))
display(ratings2_df.head(1))
```

.drop_duplicates()

```python
ratings_df = ratings_df.drop_duplicates(ignore_index=True)
print('Число строк в таблице ratings: ', ratings_df.shape[0])
```

```python
ratings_dates_df = pd.concat(
    [ratings_df, dates_df],
    axis = 1 # adding a new column from dates_df
)
display(ratings_dates_df)
```

.join()

```python
joined_false = ratings_dates_df.join(
    movies_df,
    rsuffix='_right',
    how='left'
)
display(joined_false)
```

```python
joined = ratings_dates_df.join(
    movies_df.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined.head())
```

.merge()

```python
merged =ratings_dates_df.merge(
    movies_df,
    on='movieId',
    how='left'
)
display(merged.head())
```

```python
print('Число строк в таблице ratings_dates: ', ratings_dates_df.shape[0])
print('Число строк в таблице merged: ', merged.shape[0])
print(ratings_dates_df.shape[0] == merged.shape[0])
```

```python
merge_ratings = ratings1_df.merge(ratings2_df, how='outer')
print('Число строк в таблице merge_ratings: ', merge_ratings.shape[0])
display(merge_ratings)
```
